Creating a data set utilizing the data from the 2016 and 2021 Canada cenus for possible factors due to reliability. Utilized excel in order to easily organize and visualize the data and importing it back into the code using pandas.

```python
factors = pd.read_csv("Mississauga_and_Brampton.csv")
```

```python
factors
```

Filtering each of the data into seperate variables for easy organization and understanding. 

```python
# Age Filter
# filter_columns(column) produces true or false based on whether the column heading is one of the conditions
# requires: column to be a string
def filter_columns(column):
    if (column.lower() == "date" or column.lower() == "city" or
         column.lower() == "population" or column.lower() == 'population density'
           or column.lower() == 'average household size' or column.lower() == "average income (15+)"):
        return False
    else: 
        return True
age_data = factors[filter(filter_columns, factors.columns)].copy() # filtering out only the age columns 
age_data.dropna(how='any',axis = 0, inplace=True) #dropping all empty rows

# Brampton Sorting
brampton_average_income = factors.tail(2).copy() #copying the bottom 2 rows of factors
brampton_average_income.dropna(how='all', axis=1, inplace=True) # dropping all empty columns
brampton_average_income.reset_index(drop=True, inplace=True) # resetting index

brampton_population = factors[filter(lambda col: not filter_columns(col), factors.columns)].copy() #filtering out the age columns
brampton_population = brampton_population[brampton_population["City"] == "Brampton"].head(2)  # making sure the city is only brampton
brampton_population.dropna(how='any', axis=1, inplace=True)
brampton_population.reset_index(drop=True, inplace=True)

brampton_age = age_data.tail(2)
brampton_age.reset_index(drop=True, inplace=True)

# Mississauga Sorting

mississauga_average_income = factors[(factors["City"] == "Mississauga") & 
                           ((factors["Date"] == 2015) | (factors["Date"] == 2020))].copy()
mississauga_average_income.dropna(how='all', axis=1, inplace=True)
mississauga_average_income.reset_index(drop=True, inplace=True)

mississauga_population = factors[filter(lambda column: not filter_columns(column), factors.head(2).columns)].copy()
mississauga_population = mississauga_population[mississauga_population["City"] == "Mississauga"].head(2)
mississauga_population.dropna(how='any', axis = 1, inplace=True)

mississauga_age = age_data.head(2)
```

Comparing factors and investigating if they are more significant. Due to the lack of data as the Canada Census is completed every 5 years, causation cannot be directly concluded but correlations can be highlighted. Through these, futher investigations can be sparked and additionally general trends can be spotted.

```python
plt.stairs(mississauga_age.iloc[0, :], color = "blue", label = "2016") # plotting 2016 data in blue
plt.stairs(mississauga_age.iloc[1, :], color = "red", label = "2021") # plotting 2021 data in red
plt.xlabel('Age Group')
plt.xticks(np.arange(0.5,21.5), mississauga_age.columns, rotation = 90)
plt.ylabel('Population')
plt.title('Mississauga Population Distribution by Age Group')
plt.legend()
plt.show()
```

```python
plt.stairs(brampton_age.iloc[0, :], color = "blue", label = "2016")
plt.stairs(brampton_age.iloc[1, :], color = "red", label = "2021")
plt.xlabel('Age Group')
plt.xticks(np.arange(0.5,21.5), brampton_age.columns, rotation = 90)
plt.ylabel('Population')
plt.title('Brampton Population Distribution by Age Group')
plt.legend()
plt.show()
```

Comparing the 2 graphs, a similar trend appears at the older age groups, from 50 and on. Investigating the case of the 50-54 age group more closely, there is a 13.2% increase for Brampton and a 3.3% increase for Mississauga. This does not directly correlate with the 91.9% increase in housing prices but would benefit from further investigations

Now using SciKit-Learn in order to investigate the correlation between the other factors by investigating the similaries in the slope of the linear regression as similiar - if not identical - rate of change implies a strong correlation. 

```python
from sklearn.metrics import r2_score 
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression();
# Mississauga Population
x = np.arange(0,2)
x_reshaped = x.reshape(-1, 1)
reg.fit(x_reshaped, mississauga_population["Population"])
(reg.coef_, reg.intercept_)
```

```python
r2_score(mississauga_population["Population"], reg.predict(x_reshaped))
```

```python
# Brampton Population
reg.fit(x_reshaped, brampton_population["Population"])
(reg.coef_,reg.intercept_)
```

```python
r2_score(brampton_population["Population"], reg.predict(x_reshaped))
```

```python
# Mississauga Population Density
reg.fit(x_reshaped, mississauga_population["population density"])
(reg.coef_,reg.intercept_)
```

```python
r2_score(mississauga_population["population density"], reg.predict(x_reshaped))
```

```python
# Brampton Population Density
reg.fit(x_reshaped, brampton_population["population density"])
(reg.coef_,reg.intercept_)
```

```python
r2_score(brampton_population["population density"], reg.predict(x_reshaped))
```

```python
# Mississauga Average Household Size
reg.fit(x_reshaped, mississauga_population["Average Household size"])
(reg.coef_,reg.intercept_)
```

```python
r2_score(mississauga_population["Average Household size"], reg.predict(x_reshaped))
```

```python
# Brampton Average Household Size
reg.fit(x_reshaped, brampton_population["Average Household size"])
(reg.coef_,reg.intercept_)
```

```python
r2_score(brampton_population["Average Household size"], reg.predict(x_reshaped))
```

```python
# Mississauga Average Income
reg.fit(x_reshaped, mississauga_average_income["Average income (15+)"])
(reg.coef_,reg.intercept_)
```

```python
r2_score(mississauga_average_income["Average income (15+)"], reg.predict(x_reshaped))
```

```python
# Brampton Average Income
reg.fit(x_reshaped, brampton_average_income["Average income (15+)"])
(reg.coef_,reg.intercept_)
```

```python
r2_score(brampton_average_income["Average income (15+)"], reg.predict(x_reshaped))
```

Since the data all has an r2 value of 1, it is the exact slope line that connects the points thus I can accurately compare slopes

Finding the relation of the benchmark prices

```python
housing_x = np.arange(0,69) # creating an array with the x - values for the benchmark pricing
housing_x_reshaped = housing_x.reshape(-1, 1) # reformat for Linear Regression
mississauga_y = mississauga["CompBenchmark"] # removes the index values as Linear Regression only takes in 2 values
reg.fit(housing_x_reshaped,mississauga_y)
(reg.coef_,reg.intercept_)
```

```python
r2_score(mississauga_y, reg.predict(housing_x_reshaped))
```

```python
y_2 = brampton["CompBenchmark"]
reg.fit(housing_x_reshaped,y_2)
(reg.coef_,reg.intercept_)
```

```python
r2_score(y_2, reg.predict(housing_x_reshaped))
```

The first model to represent the Mississauga housing prices has a high positive correlation so it can be a reliable predictive model to investigate future Mississauga housing prices. Although the second model for Brampton housing prices has a smaller r2 score, it still has a high positive correlation of 0.89. 


Comparing all of the slopes to all of the other calculated linear correlations, it is clear that population does not correlate with cost as the trend is negative compared to the housing prices increasing trend. Of course there could be more unreported residents within the Mississauga region however with a slope of -3638, there would have to exist a sigificant amount of unreported residents. Brampton's population does have a positive trend for both population and housing prices but it improbable for this correlation to the significant factor as it does not affect mississauga equivalently.
